```python
# read data
df_governors = pd.read_csv('./data_input/governors_county_candidate.csv')
df_president = pd.read_csv('./data_input/president_county_candidate.csv')
df_senate = pd.read_csv('./data_input/senate_county_candidate.csv')
```

```python
# We have 3112 rows and 8 columns and theres no null values, lets see the data based on the first 5 rows of total_votes by DESC. and we can remove the index column
df_president_new = df_president.sort_values(by='total_votes', ascending=False).head().reset_index(drop=True)
df_president_new
```

```python
# Now we need to check total votes group by each party
df_president.groupby('party')['total_votes'].sum().reset_index().sort_values(by='total_votes', ascending=False)
```

```python
df_president.head()
```

```python
# now we shall look by state and county and see the total votes for each party
df_presiden_vote = df_president.groupby(['state','county','party'])['total_votes'].sum().reset_index().sort_values(by='total_votes',ascending=False)
df_presiden_vote
```

```python
df_presiden_vote = df_president.groupby(['state', 'county', 'party'])['total_votes'].sum().reset_index()
df_presiden_most_vote = df_presiden_vote.set_index(['state', 'county']).sort_values(by='total_votes', ascending=False)
df_presiden_most_vote.head()
```

```python
df_presiden_state = df_presiden_vote.set_index(['state', 'county'])
df_presiden_state.head()
```

```python
#list of 3 most voted party
selected_parties = ['DEM', 'REP','LIB']
df_selected_parties = df_presiden_vote[df_presiden_vote['party'].isin(selected_parties)]
# df_selected_parties.head()

#pivot the DF
df_pivot = df_selected_parties.pivot_table(index=['state','county'], columns='party', values='total_votes')
# df_pivot.reset_index(inplace=True)
df_pivot.head()
```

```python
# Group by 'state' and 'county' and sum the total votes for each party
df_county_totals = df_selected_parties.groupby(['state', 'county', 'party'])['total_votes'].sum().unstack(fill_value=0)
df_county_totals.head()
```

```python
df_state_totals = df_selected_parties.groupby(['state', 'party'])['total_votes'].sum().unstack(fill_value=0)

# Display the DataFrame
df_state_totals.head()
```

```python
# Let's sort it from descending order based on the total votes, just to know the first state that has the most votes overall for each party 
# Group by 'state' and 'party' and sum the total votes for each party
df_state_total_votes = df_selected_parties.groupby(['state', 'party'])['total_votes'].sum().unstack(fill_value=0)

# Sort by total_votes in descending order
df_state_total_votes = df_state_total_votes.sort_values(by=['DEM', 'REP', 'LIB'], ascending=False)

# Display the DataFrame
df_state_total_votes
```

```python
# Lets see total votes for DEM based on the state
df_dem_votes = df_state_total_votes['DEM'].sort_values(ascending=False).iloc[:3]
df_dem_votes
```

```python
# Lets see total votes for LIB based on the state
df_lib_votes = df_state_total_votes['LIB'].sort_values(ascending=False).iloc[:3]
df_lib_votes
```

```python
# Lets see total votes for REP based on the state
df_rep_votes = df_state_total_votes['REP'].sort_values(ascending=False).iloc[:3]
df_rep_votes
```

```python
# We shall move to visualization
df_vis = df_presiden_vote['DEM']*100/df_presiden_vote['REP']+df_presiden_vote['DEM']
df_vis.head()
```
